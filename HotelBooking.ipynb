### Importing Libraries

```python
df = pd.read_csv('hotel_booking.csv')
```

```python
# To find the missing value in the table

df.isnull().sum()
```

```python
# To remove null column values

df.drop(['company', 'agent'], axis = 1, inplace = True)  # (inplace = True) is used to make changes in the original document and (axis = 1) for column
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# To check the variance in the values of adr by box plot

df['adr'].plot(kind = 'box')  # (adr = average daily rate) It is a columns in the table
```

```python
# There is a value 5400 which is very large value of adr with comparison to its other values. So to remove this outliner (a data point on a graph or in a set of results that is very much bigger or smaller than the next nearest data point)

df = df[df['adr'] < 5000]
```

### Data Analysis And Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)

print(cancelled_perc)

plt.figure(figsize = (5, 4))
plt.title('Reservation status count')
plt.bar(['Not canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8, 4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')  # hue means on the basis of
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1, 1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)

# The output will show 72% as not canceled and 27.9% as canceled from hotel column only in Resort Hotel.
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)

# The output will show 58% as not canceled and 41.7% as canceled from hotel column only in City Hotel.
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```
